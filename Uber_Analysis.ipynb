```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
```

```python
data_url='https://docs.google.com/spreadsheets/d/1fTpJACr1Ay6DEIgFxjFZF8LgEPiwwAFY/edit?usp=sharing&ouid=103457517634340619188&rtpof=true&sd=true'
```

```python
def load_file(data_url):
  # Modify the URL to make it accessible for pandas
  modified_url = data_url.replace('/edit?usp=sharing', '/export?format=xlsx')

  # Load ALL sheets from the Excel file
  all_sheets = pd.read_excel(modified_url, sheet_name=None)

  data={}

  # Loop through each sheet and display its first rows
  for sheet_name, df in all_sheets.items():
    data[sheet_name]=df
  return data

data=load_file(data_url)
data['Switchbacks']
```

```python
def show_metadata(data):
  lines=[sentence[0] for sentence in data['Copyright'].dropna().values.tolist()]
  text='\n'.join(lines)
  print(text)

show_metadata(data)
```

```python
def show_dictionary(data):
  headers=data['Data Dictionary'].iloc[1,:].values.tolist()
  df=data['Data Dictionary'].iloc[2:,:]
  df.columns=headers
  df.reset_index(drop=True,inplace=True)
  print(df)

show_dictionary(data)
```

```python
def load_data(data):
  display(data['Switchbacks'])

load_data(data)
```

```python
def show_time_series(data):
  df = data["Switchbacks"].copy()
  df['period_start'] = pd.to_datetime(df['period_start'])

  fig = go.Figure()

  for y_column in ['trips_pool', 'trips_express', 'rider_cancellations']:
      fig.add_trace(
          go.Scatter(
              x=df['period_start'],
              y=df[y_column],
              mode='lines+markers',
              name=y_column
          )
      )

  fig.update_layout(
      title='Time Series of Uber Trips in Boston',
      xaxis_title='Time',
      yaxis_title='Value',
      height=500
  )

  fig.show()

show_time_series(data)
```

```python
def pie_chart(data, period):
  df = data["Switchbacks"].copy()
  if period=='week':
    df['label']=df['period_start'].dt.day_name()
  if period=='month':
    df['label']=df['period_start'].dt.month_name()
  fig = px.pie(df, names='label', values='total_driver_payout')
  fig.show()

pie_chart(data, 'week')
```
